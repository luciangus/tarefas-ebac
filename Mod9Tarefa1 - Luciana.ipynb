# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc_raw.groupby('munResNome')[['IDADEMAE','IDADEPAI']].mean().round(1)
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_2002 = sinasc_raw[sinasc_raw['DTNASC']=='2019-02-20']

sinasc_2002.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().round(2)
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw['munResNome'].value_counts()
```

```python
#Ignorando o 'Município ignorado', o município com menor número de nascimentos é 'Castanheiras'
```

```python
sinasc_raw[sinasc_raw['munResNome']=='Castanheiras'][['IDADEMAE', 'IDADEPAI']].agg(['min', 'max', 'mean']).round(1)
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])

sinasc_raw[sinasc_raw['DTNASC'].dt.month == 3]['munResNome'].value_counts()
```

```python
# O município com maior número de nascimentos em março é Porto Velho
```

```python
sinasc_raw[(sinasc_raw['munResNome']=='Porto Velho') & (sinasc_raw['DTNASC'].dt.month == 3)][['QTDFILVIVO']].agg(['min', 'max', 'mean']).round(1)
```

```python
sinasc_raw[(sinasc_raw['munResNome']=='Porto Velho') & (sinasc_raw['DTNASC'].dt.month == 3)][['IDADEMAE', 'IDADEPAI']].agg(['min', 'max', 'mean']).round(1)
```

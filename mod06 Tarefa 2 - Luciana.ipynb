```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sinasc = pd.read_csv('sinasc_RO_2019.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc[['IDADEMAE']].boxplot()

plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y="PESO", data=sinasc)

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
sinasc['IDADEMAE'].plot.hist(bins=20, xlabel="Idade (anos)", ylabel="Frequência", title="Idade da Mãe")
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sinasc['PESO'].plot.hist(bins=20, xlabel="Peso (g)", ylabel="Frequência", title="Peso do Bebê")
plt.show()
```

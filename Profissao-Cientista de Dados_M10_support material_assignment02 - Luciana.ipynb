````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

previsao = pd.read_csv("previsao_de_renda.csv")
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda", 
                   data=previsao, dodge=True, errorbar=('ci', 90))
````

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda", 
                   data=previsao, dodge=True, errorbar=('ci', 90))
````

````python
# As duas variáveis discriminam a renda, porém a posse de veículo discrimina de forma significante 
# pois os valores considerando os erros estão bastante distantes um do outro, e as barras de erro são menores.
````
